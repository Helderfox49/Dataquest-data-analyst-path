# Exploring Ebay Car Sales Data

In this project, we'll work with a dataset of used cars from eBay Kleinanzeigen, a classifieds section of the German eBay website.  

The dataset  was originally scraped and uploaded to [Kaggle]('https://www.kaggle.com/orgesleka/used-cars-database/data'). For the sake of the project few modifications have been done from the original dataset that was uploaded to Kaggle.

**The aim of this project is to clean the data and analyze the included used car listings.**


Our dataset contains 20 columns, most of which are stored as strings. There are a few columns with null values, but no columns have more than ~20% null values. There are some columns that contain dates stored as strings.

## Cleaning Columns

```python
autos.columns
```

```python
def rename_col(word):
    word = word.replace('yearOfRegistration', 'registration_year')
    word = word.replace('monthOfRegistration', 'registration_month')
    word = word.replace('notRepairedDamage', 'unrepaired_damage')
    word = word.replace('dateCreated', 'ad_created')
    word = word.replace('powerPS', 'power_ps')
    word = word.replace('nrOfPictures', 'num_photos')
    for letter in word:
        if letter.isupper():
            word = word.replace(letter, "_" + letter.lower())
    return word

new_columns = []

for col in autos.columns:
    new_columns.append(rename_col(col))
print(new_columns)

#Renaming Columns
autos.columns = new_columns
```

```python
autos.head()
```

We have made a few changes here:

Change the columns from camelcase to snakecase.  
Change a few wordings to more accurately describe the columns.  

## Initial Exploring and Cleaning

```python
autos.describe(include='all')
```

Observations.  
The *num_photos* column need to be investigated.  
The *price*, *odometer* columns are stored as text, and need to be clean.  
The *seller* and *offer_type* seem to have all apart the same values.  


```python
# let's change the price column
autos["price"].unique()
```

There are two columns, price and odometer, which are numeric values with extra characters being stored as text. We'll clean and convert these.

```python
autos["price"] = autos["price"].str.replace('$', '')
autos["price"] = autos["price"].str.replace(',', '')
autos["price"] = autos["price"].astype("int")
autos["price"].head()
```

```python
autos["odometer"].unique()
```

```python
autos["odometer"] = (autos["odometer"]
                      .str
                      .replace('km', '')
                      .str
                      .replace(',', '')
                      .astype('int')
                 )
autos.rename({"odometer": "odometer_km"}, axis=1, inplace=True)
autos["odometer_km"].head()
```

```python
autos["num_photos"].unique()
autos["num_photos"].value_counts()
```

```python
print(autos["seller"].unique(), autos["offer_type"].unique(), sep='\n')
```

The *num_photos* columns seem to have only a single values 0. This information is useless, *num_photos* columns will be dropped.  
The columns *seller* and *offer_type*  almost all of the values are the same. This is not consistent, so these columns need to be dropped.  

```python
autos = autos.drop(["num_photos", "seller", "offer_type"], axis=1)
```

### Dealing with Incorrect Registration Year Data

```python
autos["registration_year"].sort_values().unique()
```

Our intervall of acceptable values will be 1900 to 2016. Let's determine what percentage of our data has invalid values in this column.

Let's remove values that are outside this range.

```python
(~autos["registration_year"].between(1900,2016)).sum() / autos.shape[0]
```

Given that this is less than 4% of our data, we will remove these rows.

```python
autos = autos[autos["registration_year"].between(1900,2016)]
autos["registration_year"].value_counts(normalize=True).head(10)
```

It appears that most of the vehicles were first registered in the past 20 years.

## Exploring Price by Brand

```python
autos["brand"].value_counts(normalize=True)
```

There are lots of brands that don't have a significant percentage of listings, so we will limit our analysis to brands representing more than 5% of total listings.

```python
brands_counts = autos["brand"].value_counts(normalize=True)
common_brands = brands_counts[brands_counts * 100 >= 5.0].index
print(common_brands)
```

```python
brand_mean_prices = {}

for brand in common_brands:
    price_brand = autos[autos["brand"] == brand]["price"]
    mean_price = price_brand.mean()
    brand_mean_prices[brand] = round(mean_price)
    
brand_mean_prices
    
```

As recap, we can say that
- Audi, BMW and Mercedes Benz are more expensive
- Ford and Opel are less expensive
- Volkswagen is in between

## Storing Aggregate Data in a DataFrame

```python
brand_mean_mileage = {}

for brand in common_brands:
    brand_only = autos[autos["brand"] == brand]
    mean_mileage = brand_only["odometer_km"].mean()
    brand_mean_mileage[brand] = int(mean_mileage)

mean_mileage = pd.Series(brand_mean_mileage).sort_values(ascending=False)
mean_prices = pd.Series(brand_mean_prices).sort_values(ascending=False)
```

```python
brand_df = pd.DataFrame(mean_prices, columns=['mean_price'])
brand_df
```

```python
brand_df["mean_mileage"] = mean_mileage
brand_df
```

We observe that there is a slight trend to the more expensive vehicles having higher mileage, with the less expensive vehicles having lower mileage.
